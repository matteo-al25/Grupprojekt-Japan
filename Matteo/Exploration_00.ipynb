**f. utforska datan vidare med egna frågor**

```python
# checking how many lands before WWI, WWII and today (***)
years = [1912, 1936, 2016]
df[df["Year"].isin(years)].groupby("Year")["NOC"].nunique()

# df.groupby('Year')['NOC'].nunique()
```

**g. diagram över könsfördelningen**

```python
df['Sex'].value_counts().plot.pie(autopct='%1.1f%%')
plt.title('Division M/F')
plt.show()
```

**h. diagram över topp 10 länder som tagit flest medaljer**

```python
top_10 = (
    df[df["Medal"].notna()]   
    .groupby("Team")["Medal"] 
    .count()                  
    .sort_values(ascending=False) 
    .head(10)                       
)

top_10.plot.pie(autopct='%1.1f%%', title='Top 5 Countries by Medal Count')
plt.ylabel('') 
plt.show()
```

**i. plotta gärna fler saker som är intressant.**

```python
# Countries participation thru the years
summer = df[df['Season'] == 'Summer']
winter = df[df['Season'] == 'Winter']

summer_os = summer.groupby("Year")["NOC"].nunique()
winter_os = winter.groupby("Year")["NOC"].nunique()
plt.figure(figsize=(12,6))
summer_os.plot(kind='line', label='Summer Olympics')
winter_os.plot(kind='line', label='Winter Olympics')
plt.legend()
plt.show()
```
