Hur många OS har Japan deltagit i över hur många år?<br>

Hur stor andel av de Japanska medaljerna är tagna av männen och hur stor andel av de Japanska tävlarna utgör männen?<br>

```python
os_japaner = athlete_events[athlete_events['NOC'] == 'JPN']
print(f'Japan har deltagit i {os_japaner['Games'].nunique()} OS mellan {os_japaner['Year'].min()} - {os_japaner['Year'].max()}.')
print(f'Antal medaljer Japan tagit i OS: {os_japaner['Medal'].value_counts().sum()}st, varav {(os_japaner['Medal'] == 'Gold').sum()}st är guld, {(os_japaner['Medal'] == 'Silver').sum()}st är silver och {(os_japaner['Medal'] == 'Bronze').sum()}st är brons.')
japaner = os_japaner[os_japaner['Sex'] == 'M']
japanskor = os_japaner[os_japaner['Sex'] == 'F']
print(f'Männen har tagit {japaner['Medal'].value_counts().sum()}st medaljer och kvinnorna har tagit {japanskor['Medal'].value_counts().sum()}st.')
print(f'Totalt är det {os_japaner['ID'].nunique()} japaner som deltagit i OS, varav {japaner['ID'].nunique()} är män och {japanskor['ID'].nunique()} är kvinnor.')
print(f'Männen har tagit {japaner['Medal'].value_counts().sum() / os_japaner['Medal'].value_counts().sum() * 100:.2f}% av medaljerna och utgör {japaner['Name'].nunique() / os_japaner['Name'].nunique() * 100:.2f}% av de japanska atleterna.')
```

# i)

```python
placeringar = medaljer.groupby(['Year', 'NOC', 'Event', 'Medal']).size().reset_index(name='Count')
os_med_count = athlete_events.merge(placeringar, on=['Year', 'NOC', 'Event', 'Medal'], how='right')
lagsporter = placeringar[placeringar['Count'] > 1]
ej_lagsporter = placeringar[placeringar['Count'] == 1]
os_japaner_podium = os_med_count[os_med_count['NOC'] == 'JPN']
japan_lag_vinster = os_japaner_podium[os_japaner_podium['Count'] > 1]
japan_individuell_vinster = os_japaner_podium[os_japaner_podium['Count'] == 1]
os_japaner_podium.sort_values(by='Count', ascending=False)
jap_placeringar = placeringar[placeringar['NOC'] == 'JPN']
jap_placeringar_lag = jap_placeringar[jap_placeringar['Count'] > 1]
lagsporter_per_land = lagsporter.groupby('NOC')['Count'].sum().sort_values(ascending=False)
indiv_per_land = ej_lagsporter.groupby('NOC')['Count'].sum().sort_values(ascending=False)
japan_lag = japan_lag_vinster['Count'].sum()
japan_indiv = japan_individuell_vinster['Count'].sum()
```

Barplots som visar det totala antalet medaljer tagna i lagsporter gentemot individuella tävlingar för alla länder och fördelningen mellan dem för Japan.

```python
print(f'Medaljer i lagsporter: {lagsporter['Count'].sum()}, individuella medaljer: {ej_lagsporter['Count'].sum()}')

fig = plt.figure(figsize=(20, 14))
gs = gridspec.GridSpec(3, 2, height_ratios=[1, 1, 1])

ax1 = fig.add_subplot(gs[0, :])
ax1.bar(lagsporter_per_land.index, lagsporter_per_land.values)
ax1.set_title('Medaljer i lagsporter per land')
ax1.set_xlabel('Land (NOC)')
ax1.set_ylabel('Antal medaljer')
ax1.tick_params(axis='x', rotation=90)

ax2 = fig.add_subplot(gs[1, :])
ax2.bar(indiv_per_land.index, indiv_per_land.values)
ax2.set_title('Medaljer i individuella sporter per land')
ax2.set_xlabel('Land (NOC)')
ax2.tick_params(axis='x', rotation=90)

ax3 = fig.add_subplot(gs[2, 0])
ax3.bar(['Lagsporter', 'Individuellt'], [lagsporter['Count'].sum(), ej_lagsporter['Count'].sum()])
ax3.set_title('Antal medaljer i lag vs individuellt (globalt)')
ax3.set_ylabel('Antal medaljer')

ax4 = fig.add_subplot(gs[2, 1])
ax4.bar(['Lagsporter', 'Individuellt'], [japan_lag, japan_indiv])
ax4.set_title('Antal medaljer i lag vs individuellt (Japan)')
ax4.set_ylabel('Antal medaljer')

plt.tight_layout()
```

Härnäst kommer barplots som visar antalet podiumplaceringar i individuella event, alltså räknar man inte längre flera medaljer per lag utan snarare lagets placering i top 3 för varje event. Detta för att det är väldigt missvisande att räkna endast medaljer eftersom att det då ser ut som att det vunnits i lagsporter väldigt många fler gånger än i individuella tävlingar när det är motsatsen som är sant, vilket tydliggörs nedan.

```python
print(f'Globalt har det tagits medalj i {len(placeringar)} tävlingar, varav {len(lagsporter)} var lagsporter och {ej_lagsporter['Count'].sum()} var individuella tävlingar.')
print(f'Japan har tagit medalj i {len(jap_placeringar)} tävlingar, varav {len(jap_placeringar_lag)} var lagsporter och {len(japan_individuell_vinster)} var individuella tävlingar.')

fig, axes = plt.subplots(1, 2, figsize=(12, 6))

axes[0].bar(['Lagsporter', 'Individuellt'], [len(lagsporter), ej_lagsporter['Count'].sum()])
axes[0].set_title('Antal podiumplaceringar i lag vs individuellt (globalt)')
axes[0].set_ylabel('Antal podiumplaceringar')

axes[1].bar(['Lagsporter', 'Individuellt'], [len(jap_placeringar_lag), len(japan_individuell_vinster)])
axes[1].set_title('Antal podiumplaceringar i lag vs individuellt (Japan)')
axes[1].set_ylabel('Antal podiumplaceringar')

plt.tight_layout()
```

Sedan kan vi kolla på könsfördelningen av de japanska podiumplaceringarna.


```python
japan_guld = os_japaner[os_japaner['Medal'] == 'Gold']
japan_silver = os_japaner[os_japaner['Medal'] == 'Silver']
japan_brons = os_japaner[os_japaner['Medal'] == 'Bronze']
japan_guld_f = japan_guld[japan_guld['Sex'] == 'F']
japan_guld_m = japan_guld[japan_guld['Sex'] == 'M']
japan_silver_f = japan_silver[japan_silver['Sex'] == 'F']
japan_silver_m = japan_silver[japan_silver['Sex'] == 'M']
japan_brons_f = japan_brons[japan_brons['Sex'] == 'F']
japan_brons_m = japan_brons[japan_brons['Sex'] == 'M']
antal_jap = os_japaner['ID'].nunique()
antal_jap_f = japanskor['ID'].nunique()
antal_jap_m = japaner['ID'].nunique()
print(f'Japan har tagit {len(japan_guld)} guld, varav kvinnorna står för {len(japan_guld_f)}st ({len(japan_guld_f)/len(japan_guld)*100:.2f}%) och männen står för {len(japan_guld_m)}st ({len(japan_guld_m)/len(japan_guld)*100:.2f}%).')
print(f'Japan har tagit {len(japan_silver)} silver, varav kvinnorna står för {len(japan_silver_f)}st ({len(japan_silver_f)/len(japan_silver)*100:.2f}%) och männen står för {len(japan_silver_m)}st ({len(japan_silver_m)/len(japan_silver)*100:.2f}%).')
print(f'Japan har tagit {len(japan_brons)} brons, varav kvinnorna står för {len(japan_brons_f)}st ({len(japan_brons_f)/len(japan_brons)*100:.2f}%) och männen står för {len(japan_brons_m)}st ({len(japan_brons_m)/len(japan_brons)*100:.2f}%).')
print(f'Totalt är det {antal_jap} japaner som deltagit i OS, varav {antal_jap_f} är kvinnor ({antal_jap_f/antal_jap*100:.2f}%) och {antal_jap_m} är män ({antal_jap_m/antal_jap*100:.2f}%).')
print(f'Kvinnorna ({antal_jap_f/antal_jap*100:.2f}%) har tagit {japaner['Medal'].value_counts().sum()/os_japaner['Medal'].value_counts().sum()*100:.2f}% av medaljerna och männen ({antal_jap_m/antal_jap*100:.2f}%) har tagit {japaner['Medal'].value_counts().sum()/os_japaner['Medal'].value_counts().sum()*100:.2f}% av medaljerna')
```

```python
fig, axes = plt.subplots(2, 2, figsize=(12, 12))

axes[0, 0].pie([len(japan_guld_f), len(japan_guld_m)], labels=['Kvinnor', 'Män'], autopct='%2.2f%%', startangle=90)
axes[0, 0].set_title('Könsfördelning guldplaceringar (Japan)')

axes[0, 1].pie([len(japan_silver_f), len(japan_silver_m)], labels=['Kvinnor', 'Män'], autopct='%2.2f%%', startangle=90)
axes[0, 1].set_title('Könsfördelning silverplaceringar (Japan)')

axes[1, 0].pie([len(japan_brons_f), len(japan_brons_m)], labels=['Kvinnor', 'Män'], autopct='%2.2f%%', startangle=90)
axes[1, 0].set_title('Könsfördelning bronsplaceringar (Japan)')

axes[1, 1].pie([antal_jap_f, antal_jap_m], labels=['Kvinnor', 'Män'], autopct='%2.2f%%', startangle=90)
axes[1, 1].set_title('Könsfördelning atleter (Japan)')

plt.tight_layout()
```

Här ser vi då att männen är överrepresenterade när det kommer till guldplaceringar men underrepresenterade i silver och brons, samt medaljer överlag då det är fler silver och brons tagna. Av detta kan man spekulera att männen överlag har en högre högstanivå medans kvinnorna håller en jämnare hög nivå.
