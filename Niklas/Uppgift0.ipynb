# f)

Hur många OS har Japan deltagit i?<br>
Hur många medaljer har Japan tagit i dessa OS?<br>
Hur ser könsfördelningen ut för dessa medaljer?<br>
Hur många individuella tävlare har Japan haft och hur är könsfördelningen mellan dem?<br>
Hur stor andel av de Japanska medaljerna är tagna av männen och hur stor andel av de Japanska tävlarna utgör männen?<br>

```python
os_japaner = athlete_events[athlete_events['NOC'] == 'JPN']
print(f'Antal OS Japan deltagit i: {os_japaner['Games'].nunique()}')
print(f'Antal medaljer Japan tagit i OS: {os_japaner['Medal'].value_counts().sum()}st, varav {(os_japaner['Medal'] == 'Gold').sum()}st är guld, {(os_japaner['Medal'] == 'Silver').sum()}st är silver och {(os_japaner['Medal'] == 'Bronze').sum()}st är brons.')
japaner = os_japaner[os_japaner['Sex'] == 'M']
japanskor = os_japaner[os_japaner['Sex'] == 'F']
print(f'Männen har tagit {japaner['Medal'].value_counts().sum()}st medaljer och kvinnorna har tagit {japanskor['Medal'].value_counts().sum()}st.')
print(f'Totalt är det {os_japaner['Name'].nunique()} japaner som deltagit i OS, varav {japaner['Name'].nunique()} är män och {japanskor['Name'].nunique()} är kvinnor.')
print(f'Männen har tagit {japaner['Medal'].value_counts().sum() / os_japaner['Medal'].value_counts().sum() * 100:.2f}% av medaljerna och utgör {japaner['Name'].nunique() / os_japaner['Name'].nunique() * 100:.2f}% av de japanska tävlarna.')
```

# g)

```python
fig, axes = plt.subplots(1, 2, figsize=(12, 5))
athlete_events['Sex'].value_counts().plot(kind='bar', ax=axes[0])
axes[0].set_title('Könsfördelning')
axes[0].set_xlabel('Kön')
axes[0].set_ylabel('Antal')
axes[0].tick_params(axis='x', rotation=0)
os_japaner['Sex'].value_counts().plot(kind='bar', ax=axes[1])
axes[1].set_title('Könsfördelning för Japaner')
axes[1].set_xlabel('Kön')
axes[1].set_ylabel('Antal')
axes[1].tick_params(axis='x', rotation=0)
```

# h)

```python
medaljer = athlete_events[athlete_events['Medal'].notna()]
medaljer['NOC'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 medaljplaceringar')
plt.xlabel('Land')
plt.ylabel('Antal medaljer')
```

# i)

```python
placeringar = medaljer.groupby(['Year', 'NOC', 'Event', 'Medal']).size().reset_index(name='Count')
os_med_count = athlete_events.merge(placeringar, on=['Year', 'NOC', 'Event', 'Medal'], how='right')
lagsporter = placeringar[placeringar['Count'] > 1]
ej_lagsporter = placeringar[placeringar['Count'] == 1]
print(f'Medaljer i lagsporter: {lagsporter['Count'].sum()}, individuella medaljer: {ej_lagsporter['Count'].sum()}')
print(f'Podiumplaceringar för lagsporter: {len(lagsporter)}, genomsnittlig lagstorlek: {lagsporter['Count'].sum() / len(lagsporter)}')
os_japaner_podium = os_med_count[os_med_count['NOC'] == 'JPN']
japan_lag_vinster = os_japaner_podium[os_japaner_podium['Count'] > 1]
japan_individuell_vinster = os_japaner_podium[os_japaner_podium['Count'] == 1]
os_japaner_podium.sort_values(by='Count', ascending=False)
jap_placeringar = placeringar[placeringar['NOC'] == 'JPN']
jap_placeringar_lag = jap_placeringar[jap_placeringar['Count'] > 1]
print(f'Japan har tagit medalj i {len(jap_placeringar)} tävlingar, varav {len(jap_placeringar_lag)} var lagsporter och {len(japan_individuell_vinster)} var individuella sporter.')
```
